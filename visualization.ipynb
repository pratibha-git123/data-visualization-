```python
df = pd.read_csv('practice.csv')
df
```

```python
x = df['Name']
y = df['Marks']
plt.plot(x,y, marker='*', color='y', lw=2)#LW Afor line width
#plt.plot(x,y, marker='D')
#plt.plot(x,y, marker='o')
#plt.plot(x,y, '<:r')
#plt.plot(x,y, marker='D')
```

```python
plt.plot(x,y, '*r' , ms=18)# ms for marker size
```

```python
df.head()
```

```python
plt.plot(x,y, '*-b' )
plt.title('ADIT - Class',{'color':'red', 'family':'serif'})
plt.grid( linestyle='-.',color='g')
#plt.grid(axis='y')for only y 
```

```python
plt.plot(x,y, 'o-.g' )
plt.title('ADIT - Class',{'color':'b', 'family':'serif'})
plt.grid( linestyle='-.',color='r', mouseover=True)
```

```python
plt.figure()
plt.title('ADIT - Class',{'color':'red', 'family':'serif'})
plt.grid( linestyle='-.',color='r')
plt.bar(x,y, color='blue')
```
