```python
import seaborn as sns
import pandas as pd
```

```python
df= sns.load_dataset('tips')
df.head()
```

```python
sns.scatterplot(x="total_bill",y="tip", data=df ,hue="day")
```

```python
sns.scatterplot(x="total_bill",y="tip", data=df ,hue="time")
```

```python
sns.scatterplot(x="total_bill",y="tip", data=df ,hue="smoker")
```

```python
sns.scatterplot(x="total_bill",
                y="tip", 
                data=df ,
                hue="day",
                size="size",
                palette="Accent_r",
                sizes=(10,100)
               )
```

```python
sns.lineplot(x="total_bill",y="tip",data=df,hue="sex",style="smoker",markers=True,palette="Paired")
```

```python
sns.barplot(x='size',y='total_bill',data=df)
```

```python
sns.barplot(x='size',y='total_bill',data=df,hue='sex',palette='RdYlGn_r')
```

```python
sns.histplot(x='size', data=df, bins=50)
```

```python
sns.kdeptot(x='size', data=df, hue='smoker',fill=True, linewidth=2.5)
```
